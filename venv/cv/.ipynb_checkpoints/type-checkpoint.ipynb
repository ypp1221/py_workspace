```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../image/houzi.jpg',1)
print img.shape
#使用np数组生图片类型
img1 = np.zeros(img.shape,np.uint8)
img1 = cv.bitwise_not(img1)
print img1.shape
plt.subplot(221),plt.imshow(img),plt.title('img')
plt.subplot(222),plt.imshow(img1),plt.title('img1')
plt.show()
```
