```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#获得图像的均值和方差
def get_mean(img,img1):
    m = cv.mean(img)
    print 'm mean is '+ str(m)
    m1 = cv.mean(img1)
    print 'm1 mean is '+ str(m1)
#获得图像的均值和方差
def get_std(img,img1):
    m = cv.meanStdDev(img)
    m1 = cv.meanStdDev(img1)
    print 'm std is '+ str(m)
    print 'm1 std is '+ str(m1)
img = cv.imread('../image/houzi.jpg',1)
m,n = img.shape[:2]
img1 = np.zeros((m,n),np.uint8)
get_mean(img,img1)
get_std(img,img1)
plt.subplot(221),plt.imshow(img),plt.title('oriangle')
plt.subplot(222),plt.imshow(img1),plt.title('zero')

plt.show()
```
