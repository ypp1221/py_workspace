```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../image/houzi.jpg',1)
```

```python
#Making borders for images(Padding)
bule = [255,0,0]
#top,bottom,left,right-border width in number of pixels im correspondig directions
replicate = cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_CONSTANT,value = bule)


plt.subplot(221),plt.imshow(replicate)
plt.subplot(222),plt.imshow(img)
plt.show()
```
